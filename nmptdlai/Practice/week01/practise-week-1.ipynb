#PHAN RA COT

pd.melt()là một hàm trong thư viện Pandas được sử dụng để chuyển đổi hoặc định hình lại dữ liệu từ định dạng rộng sang định dạng dài. Hàm này hữu ích khi bạn cần "bỏ trục" một DataFrame, chuyển các cột thành hàng.

Cú pháp:
    pd.melt(frame, id_vars=None, value_vars=None, var_name=None, value_name='value', col_level=None)

Các thông số:

frame : DataFrame mà bạn muốn định hình lại.

id_vars : Các cột được giữ làm định danh (các cột này sẽ không thay đổi). Đây là các cột "cố định" sẽ không bị thay đổi.

value_vars : Các cột bạn muốn bỏ trục. Các cột này sẽ được chuyển thành hàng.

var_name : Tên đặt cho cột mới sẽ chứa tên các cột trước đó (từ value_vars). Nếu Không có, tên mặc định sẽ là variable.

value_name : Tên đặt cho cột mới sẽ chứa các giá trị từ các cột không có trục. Theo mặc định, đây là value.

col_level : Nếu các cột có nhiều cấp, điều này sẽ chỉ định cấp nào sẽ bỏ trục.


```python
df = pd.melt(df, id_vars=['Id', 'Age', 'Weight_kgs', 'Firstname', 'Lastname'], value_name='PulseRate', var_name='gender_time').sort_values(['Id', 'Age', 'Weight_kgs', 'Firstname', 'Lastname'])
```

```python
print(df)
```

```python
df_temp = df['gender_time'].str.extract("(\D)(\d+)(\d{2})", expand=True)
print(df_temp)
```

```python
df_temp.columns=['Gender', 'Lower_hour', 'Upper_hour']
```

```python
df = pd.concat([df, df_temp], axis=1)
print(df)
```

5. Vẽ biểu đồ cột so sánh nhịp tim trung bình theo từng nhóm giờ 00-06, 06-12,12-18 có 
phân loại theo giới tính

```python
bins = [0, 6, 12, 18]
labels = ['00-06', '06-12', '12-18']
df['Time Group'] = pd.cut(df['Lower_hour'], bins=bins, labels=labels, right=False)

average_pulse_by_time_gender = df.groupby(['Time Group', 'Gender'])['PulseRate'].mean().unstack()

plt.figure(figsize=(10, 6))
average_pulse_by_time_gender.plot(kind='bar')
```

6. Vẽ biểu đồ phân tán thể hiện tương quan giữa trung bình độ tuổi và trung bình nhịp tim

```python
mean_age = df.groupby('Age').mean()
mean_pulse = df.groupby('Age')['PulseRate'].mean().reset_index()
print(mean_pulse)

fig, ax = plt.subplots()
img = ax.scatter(mean_pulse['Age'], mean_pulse['PulseRate'], c=mean_pulse['PulseRate'], alpha=0.7)
fig.colorbar(img)
```

7. Vẽ biểu đồ đường thể hiện sự thay đổi của nhịp tim trung bình các bệnh nhân theo thứ tự
khung giờ 00-06, 06-12,12-18 theo từng nhóm giới tính.

```python
bins = [0, 6, 12, 18]
labels = ['00-06', '06-12', '12-18']

df['Time Group'] = pd.cut(df['Lower_hour'], bins=bins, labels=labels, right=False)
# print(df['Time Group'])

mean_pulse_by_time_gender = df.groupby(['Time Group', 'Gender'])['PulseRate'].mean().reset_index()
print(mean_pulse_by_time_gender)

for gender in mean_pulse_by_time_gender['Gender'].unique():
    # print(mean_pulse_by_time_gender['Gender'] == gender)
    gender_data = mean_pulse_by_time_gender[mean_pulse_by_time_gender['Gender'] == gender]
    # print(gender_data)
    gender_data['Time Group'] = gender_data['Time Group'].astype(str)
    plt.plot(gender_data['Time Group'], gender_data['PulseRate'], marker='o', label= 'Nu' if gender == 'f' else 'Nam')

plt.legend(loc='lower left')
```
