# Hperparameter tuning model with Grid Search
# Su dung do do danh gia accuracy tren tap train va test de so sanh
# Tra loi cau hoi lua chon K bao nhieu la tot nhat cho mo hinh K-NN

```python
# n_neighbors: Numer of neighbors to use, default is 5
# algorithm
# metric: distance computation, default is "minkowski", 
```

```python
from sklearn.model_selection import ParameterGrid

# Define the parameter grid
param_grid = {
    'n_neighbors': range(2, 7),
    'algorithm': ['ball_tree', 'kd_tree'],
    'metric': ['euclidean','minkowski']
}

param_candidates = ParameterGrid(param_grid)

print(f'Number of candidates: {len(param_candidates)}')

res = []

for i, params in enumerate(param_candidates):
    model = KNeighborsClassifier().set_params(**params)
    model.fit(x_train, y_train)
    score_train = model.score(x_train, y_train)
    score_test = model.score(x_test, y_test)
    res.append((params, score_train, score_test))
    print(f'{i}/{len(param_candidates)-1}:', params, score_train, score_test)

print(res)

# Find the maximum scores
max_score_train = max(res, key=lambda x: x[1])
max_score_test = max(res, key=lambda x: x[2])

print(f'Max score on train: {max_score_train}')
print(f'Max score on test: {max_score_test}')

list_index_on_res = range(len(res))
list_score_on_train = [x[1] for x in res]
print(list_score_on_train)

list_score_on_test = [x[2] for x in res]

print(list_score_on_test)

# Generate plot
#plt.figure(figsize=(50,20))
plt.plot(list_index_on_res, list_score_on_train, label = 'Training dataset Accuracy')
plt.plot(list_index_on_res, list_score_on_test, label = 'Testing dataset Accuracy')
plt.legend()
plt.xlabel('Grid params (n_neighbors, algorithm, metric)')
plt.xticks(range(len(res)), range(len(res)))
plt.ylabel('Accuracy')
plt.show()
```

```python
#5.2 Sử dụng tham số K để đo giá trị K tốt nhất

import numpy as np
neighbors = np.arange(1, 25)
train_accuracy = np.empty(len(neighbors))
test_accuracy = np.empty(len(neighbors))


# Loop over K values
for i, k in enumerate(neighbors):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(x_train, y_train)

    # Compute training and test data accuracy
    train_accuracy[i] = knn.score(x_train, y_train)
    test_accuracy[i] = knn.score(x_test, y_test)

    print(f'k={k}, train score={train_accuracy[i]}, test_score={test_accuracy[i]}')

# Generate plot
plt.plot(neighbors, train_accuracy, label='Training dataset Accuracy')
plt.plot(neighbors, test_accuracy, label='Testing dataset Accuracy')
plt.xticks(np.arange(1, 25), np.arange(1, 25))
plt.legend()
plt.xlabel('n_neighbors')
plt.ylabel('Accuracy')
plt.show()
```

```python
#20. Đánh giá mô hình thông qua đồ thị AUC & ROC
#ROC, AUC (cách đơn giản)
X_test = x_test
from sklearn.metrics import roc_curve, auc

y_score = knn.predict_proba(X_test)[:,1]
fpr, tpr, _ = roc_curve(y_test, y_score)

plt.title('ROC curve')
plt.xlabel('FPR (Precision)')
plt.ylabel('TPR (Recall)')

plt.plot(fpr, tpr)
plt.plot((0,1), ls='dashed', color='black')
plt.show()
print('Area under curve (AUC): ', auc(fpr, tpr))
```

```python
#ROC, AUC (cách nâng cao)

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import metrics

y_pred_proba = knn.predict_proba(X_test)[:,1]
fpr, tpr, _ = metrics.roc_curve(y_test, y_pred_proba)
auc = metrics.roc_auc_score(y_test, y_pred_proba)

plt.plot(fpr, tpr, 'go-', label="AUC=" + str(auc))
plt.plot([0,1], [0,1], '--')

plt.title("AUC & ROC Curve")
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend(loc=4)
plt.fill_between(fpr, tpr, facecolor='lightgreen', alpha=0.7)
plt.show()
```
