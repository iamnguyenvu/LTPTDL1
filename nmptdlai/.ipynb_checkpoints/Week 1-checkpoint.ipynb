# DATA FRAME MANIPULATION

```python
df['M1'].loc[::-5]
df.loc[55:77, ['M1', 'M2']]
df.iloc[:-5]
df.sort_values(['M1', 'M2'], ascending=[True, True])
df.loc[df['M1'] == 3.0]
df[(df['M1'] > 5) & (df['M2'] > 5)]
```

Mức độ dễ
1. Hiểu và truy xuất dữ liệu
Đọc dữ liệu từ tệp CSV và hiển thị 5 dòng đầu tiên.
Lấy thông tin về các cột trong DataFrame, bao gồm tên các cột và kiểu dữ liệu của chúng.
Kiểm tra xem có bao nhiêu giá trị null trong mỗi cột.
Hiển thị các giá trị duy nhất trong cột 'GT' (giới tính).
Hiển thị tất cả các sinh viên trong khu vực 'KV' là 'A1'.
2. Lọc dữ liệu cơ bản
Lọc ra các sinh viên có điểm 'M1' lớn hơn 6.
Lọc các sinh viên có điểm 'M1' nhỏ hơn hoặc bằng 5 và chỉ lấy các cột 'M1', 'M2'.
Lọc và hiển thị các sinh viên có 'GT' là 'Nam' và có điểm 'M1' từ 7 trở lên.
Sắp xếp DataFrame theo cột 'M2' theo thứ tự giảm dần.
3. Manipulate DataFrame
Thêm một cột mới 'M1_M2_Sum' là tổng của cột 'M1' và 'M2'.
Đổi tên cột 'NGHE' thành 'Ngành nghề'.
Xóa cột 'NGHE' khỏi DataFrame.

```python
# df.info()
df['GT']
df[df['M1'] <= 5][['M1', 'M2']]
df[(df['GT'] == 'M') & (df['M1'] > 6)]
df.sort_values(['M2'], ascending=False)
df[df['M1'] <= 5][['M1', 'M2']]
```

4. Nhóm và tính toán thống kê
Nhóm dữ liệu theo 'KV' và tính tổng điểm thi 'M1' cho mỗi khu vực.
Nhóm dữ liệu theo 'KV' và tính giá trị trung bình của các cột 'M1', 'M2' cho mỗi nhóm.
Tính độ lệch chuẩn của cột 'M1' cho từng khu vực 'KV'.
Tính số lượng sinh viên trong mỗi khu vực 'KV' bằng cách sử dụng groupby và count().
5. Tạo bảng pivot
Tạo một bảng pivot để hiển thị giá trị 'M1' và 'M2' cho từng khu vực 'KV' với các hàm tổng hợp min, mean, và max.
Tạo một bảng pivot để hiển thị giá trị 'M1' và 'M2' cho các khu vực 'KV' và các nhóm 'KT'.
6. Lọc dữ liệu theo điều kiện phức tạp
Lọc các sinh viên có điểm 'M1' lớn hơn 6 và điểm 'M2' lớn hơn 5.
Lọc các sinh viên có 'GT' là 'Nam', khu vực 'KV' là 'A1' và điểm 'M1' lớn hơn hoặc bằng 7.

```python
df.groupby(['KV'])['M1'].agg(['sum'])
df.groupby(['KV'])['M1', 'M2'].agg('mean')
df.groupby(['KV'])['M1'].agg('std')
df.groupby(['KV'])['M1'].agg('count')
pd.pivot_table(df, values=['M1', 'M2'], columns='KV', aggfunc=['min', 'mean', 'max'])
pd.pivot_table(df, values=['M1', 'M2'], index='KT', columns=['KV'], aggfunc=['min', 'mean', 'max'])
df[(df['M1'] > 6) & (df['M2'] > 5)]
df[(df['GT'] == 'M') & (df['KV'] == '1') & (df['M1'] >= 5)]
```

7. Tính toán thống kê nâng cao
Tính toán giá trị trung bình của cột 'M1' cho mỗi khu vực 'KV', nhưng chỉ tính cho các sinh viên có điểm 'M1' lớn hơn 5.
Sử dụng aggregate() để tính toán giá trị tối thiểu, tối đa và trung bình cho các cột 'M1', 'M2' và 'M3' theo từng khu vực 'KV'.
Tính toán giá trị trung bình của các cột 'M1' và 'M2' cho từng nhóm 'KV' và 'KT'.
8. Sử dụng pivot_table với nhiều hàm tổng hợp
Tạo bảng pivot để hiển thị các giá trị 'M1', 'M2' theo từng khu vực 'KV' và nhóm 'KT' với các hàm tổng hợp min, mean, std, và sum.
Sử dụng pivot_table để tính toán giá trị trung bình và tổng điểm 'M1', 'M2' cho mỗi khu vực 'KV' theo từng nhóm 'NGHE'.
9. Tính toán theo điều kiện phức tạp
Lọc dữ liệu cho những sinh viên có điểm thi 'M1', 'M2', 'M3' đều lớn hơn 6, và sau đó tính toán giá trị trung bình của 'M1', 'M2' và 'M3' cho nhóm này.
Lọc và tính toán độ lệch chuẩn của 'M1' cho các sinh viên có điểm 'M1' lớn hơn 6 và 'M2' lớn hơn 5.

```python
df[df['M1'] > 5].groupby(['KV'])['M1'].agg(['min', 'mean', 'max'])
df.groupby(['KV'])['M1', 'M2', 'M3'].agg(['min', 'mean', 'max'])
df.groupby(['KV', 'KT'])['M1', 'M2'].agg('mean')

pd.pivot_table(df, values=['M1', 'M2'], index=['KT'], columns='KV', aggfunc=['min', 'mean', 'std', 'sum'])
pd.pivot_table(df, values=['M1', 'M2'], index='NGHE', columns='KV', aggfunc=['mean', 'sum'])

df[(df['M1'] > 5) & (df['M2'] > 5) & (df['M3'] > 5)][['M1', 'M2', 'M3']].agg(['mean'])
df[(df['M1'] > 5) & (df['M2'] > 5)]['M1'].agg('std')
```

10. Xử lý dữ liệu thiếu và thay thế giá trị
Kiểm tra các giá trị thiếu trong DataFrame và thay thế các giá trị thiếu trong cột 'DT' bằng giá trị trung bình của cột này.
Thay thế giá trị thiếu trong các cột số học (ví dụ: 'M1', 'M2') bằng giá trị trung bình của mỗi cột.
11. Tạo và phân tích cột mới
Tạo một cột 'Điểm Trung Bình' là trung bình của các cột 'M1', 'M2', 'M3'. Sau đó phân loại sinh viên vào các nhóm 'Khá', 'Giỏi', 'Trung bình' dựa trên giá trị trung bình của họ.
Tạo một cột 'Năm Sinh' cho sinh viên dựa trên năm hiện tại (2024), giả sử tuổi trung bình của sinh viên là 20.
12. Kết hợp dữ liệu với nhiều bảng (merge)
Nếu có một bảng dữ liệu khác chứa thông tin chi tiết về sinh viên (ví dụ: bảng thông tin gia đình, hoặc bảng điểm), thực hiện phép kết hợp (merge) giữa hai bảng này dựa trên cột 'MSSV' và hiển thị kết quả.
13. Tính toán và phân tích nâng cao
Sử dụng phương pháp groupby() kết hợp với agg() để tính toán các thống kê phức tạp cho các nhóm sinh viên, ví dụ như tổng điểm của các nhóm sinh viên theo giới tính và khu vực, sau đó phân tích sự phân bố điểm trong từng nhóm.
